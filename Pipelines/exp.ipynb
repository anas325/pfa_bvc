```python
import pandas as pd
```

```python
df = pd.read_json('data/lematin_companies.json')
df.drop(['cours','variation'], axis=1).to_csv('data/companies.csv', index=False)
```

```python
from db.db import get_driver
```

```python
driver = get_driver()
```

```python
rows = df.to_dict("records")

query = """
UNWIND $rows AS row
MERGE (c:Company {ticker: row.ticker})
SET c.libelle = row.libelle,
	c.secteur = row.secteur,
	c.siege_social = row.siege_social,
	c.cours = row.cours,
	c.variation = row.variation
"""

with driver.session() as session:
	session.run(query, rows=rows)
	
```
